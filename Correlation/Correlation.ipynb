```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
from scipy import stats
import requests
import time
import os
from scipy.stats import linregress
```

```python
#Output File (CSV)
data_price= "data/US_Price.csv"
data_rig_count = "data/NG_Rig_Count.csv"
data_production = "data/NG_Production1.csv"
data_storage = "data/US_Storage.csv"
data_exports ="data/NG_Export.csv"
data_imports = "data/NG_Import.csv"
data_consumption = "data/NG_Consumption.csv"
data_montly_price = "data/NG_Monthly_Price.csv"
```

```python
#reading files
US_Price= pd.read_csv(data_price)
US_Montly_Price = pd.read_csv(data_montly_price)
US_Rig_Count = pd.read_csv(data_rig_count)
NG_Production1 = pd.read_csv(data_production)
US_Storage = pd.read_csv(data_storage)
US_Exports = pd.read_csv(data_exports)
US_Imports =pd.read_csv(data_imports)
US_Consumptions = pd.read_csv(data_consumption)
```

```python
#1
# Finding NG Prices vs NG rig count
#rename the column 
US_Rig_Count.rename(columns={'Date' : 'Week'},inplace=True)

#merg two data set
merge_Price_rigcount = pd.merge(US_Price, US_Rig_Count, on= 'Week')

#Change data set into integer
merge_Price_rigcount['Oil'] = pd.to_numeric(merge_Price_rigcount['Oil'], errors='coerce').fillna(0).astype(np.int64)
```

```python
#What is the relationship between NG price and NG rig count
x= merge_Price_rigcount['Price']
y = merge_Price_rigcount['Gas']

vc_slope, vc_int, vc_r, vc_p, vc_std_err = stats.linregress(x, y)
vc_fit = vc_slope * x + vc_int
plt.scatter(x, y)
plt.xticks(x, rotation=90,fontsize=4)
plt.xlabel('US Natural Gas Price')
plt.ylabel('US Rig Count(Gas)')
plt.title("US Gas Price Vs US Gas Rig Count")
plt.savefig('images/Price_vs_Rig Count.png')
plt.show()
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
print(f'Sample size: {y.count()}. The correlation coefficient:{round(st.pearsonr(x,y)[0],2)}, the {stats.linregress(x,y)}, r-squared: {r_value**2}')
```

```python
#2
#for correlation between NG Price and US Storage
# Merge two dataframes using an inner join
merge_Price_Storage = pd.merge(US_Price, US_Storage, on="Week")
```

```python
#What is the relationship between NG price and NG Storage
x2= merge_Price_Storage['Price']
y2 = merge_Price_Storage['Storage']
vc_slope, vc_int, vc_r, vc_p, vc_std_err = stats.linregress(x2, y2)
vc_fit = vc_slope * x2 + vc_int

plt.scatter(x2, y2)
plt.plot(x2,vc_fit,"--")
plt.xticks(x2, rotation=90)
plt.xlabel('US Natural Gas Prices')
plt.ylabel('US Storage')
plt.title("US Price Vs US Storage")
plt.savefig('images/price_vs_storage.png')
plt.show()
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(x2,y2)
print(f'Sample size: {y2.count()}. The correlation coefficient:{round(st.pearsonr(x2,y2)[0],2)}, the {stats.linregress(x2,y2)}, r-squared: {r_value**2}')
```

```python
#3
# for NG prices vs NG demand(million) + export(millions)

#merge two data set with inner join 
merge_exports_consumption= pd.merge (US_Consumptions, US_Exports, on= 'Month')

#sum of consumption and exports
sum_consumption_exports= merge_exports_consumption['Consumption'] + merge_exports_consumption['Export']
merge_exports_consumption['Consumption + Exports'] = sum_consumption_exports
```

```python
#merging data for correlation
merge_price_consumption = pd.merge(US_Montly_Price, merge_exports_consumption, on = 'Month')
```

```python
#What is the relationship between NG price and NG consumption + Exports
x3= merge_price_consumption['Price']
y3 = merge_price_consumption['Consumption + Exports']

vc_slope, vc_int, vc_r, vc_p, vc_std_err = stats.linregress(x3, y3)
vc_fit = vc_slope * x3 + vc_int

plt.scatter(x3,y3)
plt.plot(x3,vc_fit,"--")
plt.xticks(x3, rotation=90)
plt.xlabel('US price')
plt.ylabel('US Comsumption')
plt.title("Price vs Consumption + Exports")
plt.savefig('images/Price_Consumption+Exports.png')
plt.show()
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(x3,y3)
print(f'Sample size: {y3.count()}. The correlation coefficient:{round(st.pearsonr(x3,y3)[0],2)}, the {stats.linregress(x3,y3)}, r-squared: {r_value**2}')
```

```python
#4
#for NG Prices vs Production(millions) + import (millions)

#merge the dataset based on inner join
merge_production_import= pd.merge(NG_Production1, US_Imports, on = 'Month')

#add two column
sum_production_import = merge_production_import['Production'] + merge_production_import['Import']
merge_production_import['Production + Import'] = sum_production_import
```

```python
#merge the data (final)
merge_production_price =pd.merge(merge_production_import, US_Montly_Price, on ='Month')
```

```python
#What is the relationship between US price vs production + Imports
x4= merge_production_price['Price']
y4 = merge_production_price['Production + Import']

vc_slope, vc_int, vc_r, vc_p, vc_std_err = stats.linregress(x4, y4)
vc_fit = vc_slope * x4 + vc_int

plt.scatter(x4,y4)
plt.plot(x4,vc_fit,"--")
plt.xticks(x4, rotation=90)
plt.xlabel('US price')
plt.ylabel('US Production')
plt.title("Price vs Production + Imports")
plt.savefig('images/price_vs_production+Imports.png')
plt.show()
```
